Glyphs in Bokeh are the visual elements that represent data in the plot. These are the basic building blocks for creating Bokeh plots. Examples of glyphs include lines, circles, rectangles, and text.

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

x = [1, 2, 3, 4, 5]
y = [2, 5, 3, 7, 1]
size = [10, 20, 15, 30, 25]
color = ['red', 'green', 'blue', 'orange', 'purple']

source = ColumnDataSource(data=dict(x=x, y=y, size=size, color=color))

p = figure()

p.circle(x='x', y='y', size='size', color='color', source=source)

show(p)
```

Bokeh server is a Python server that allows for the creation of interactive web applications and real-time data visualization. With the Bokeh server, users can create web applications that update dynamically in response to changes in data, user input, or other events.

To use the Bokeh server, you first need to create a Bokeh plot or application using the Bokeh library in Python. Once you have created your plot or application, you can then run it on the Bokeh server using the bokeh serve command. This will start the Bokeh server and make your plot or application available through a web browser.

When running your plot or application on the Bokeh server, you can use the Bokeh API to update the plot or application in real-time based on user input or other events. For example, you might create a web application that allows users to explore a dataset by selecting different variables to plot or filtering the data based on different criteria. As users make selections or changes, the application would update the plot in real-time to reflect the new data or settings.

Overall, the Bokeh server provides a powerful tool for creating interactive data visualizations and web applications that can be easily shared and updated in real-time

```python
from bokeh.plotting import figure
from bokeh.embed import components

p = figure(plot_width=400, plot_height=400)
p.circle([1, 2, 3], [4, 5, 6])
script, div = components(p)

from bokeh.plotting import save
save(p, 'plot.html')
```

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route("/")
def plot():
    return render_template('plot.html', script=script, div=div)

if __name__ == "__main__":
    app.run()
```
